```python
import pandas as pd
import seaborn as sns
```

# Data Import

```python
import_db = pd.read_csv('Customer-Churn-Records.csv')
import_db.head()
```

```python
import_db.info()
```

```python
import_db.describe(include='all')
```

```python
import_db.columns
```

```python
# Droping columns that won't be used
# RowNumber, CustomerID, Surname, Complain

# Complain was dropped for the sake of differentiating the machine learning models, because when not dropped all the models have had very similar accuracy. 
# So to create a scenario where a better model exists, this was necessary.

import_db = import_db.drop(['RowNumber', 'CustomerId', 'Surname', 'Complain'],axis=1)
import_db.head()
```

```python
import_db.columns
```

```python
# Encoding categorial Data
# Categorical data represents distinct, non-numeric categories (e.g., Geography, Card Type, Gender). 
# Encoding converts it into numeric format for machine learning compatibility.

import_db = pd.get_dummies(import_db,drop_first=True)
import_db.head()
```

```python
# Understanding data distribution
import_db['Exited'].value_counts()
```

```python
sns.countplot(x='Exited', data=import_db)
# With this graph we can see that the dataset is unbalanced, having more 0s than 1s.
```

```python
# Separating Features (X) and Labels (y)
# Features (X): The input variables or attributes that are used to make predictions.
# Labels (y): The output or target variable that the model aims to predict.

X = import_db.drop('Exited',axis=1)
y = import_db['Exited']
```

```python
# Splitting dataset
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=42,stratify=y)
```

```python
# Feature Scalling
# Standardize features for consistent scale; vital for algorithms sensitive to scale differences.
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()                 # Create a StandardScaler object
X_train = sc.fit_transform(X_train)   # Fit on training data and transform features
X_test = sc.transform(X_test)         # Transform test features using the same scale
```

```python
 # Logistic Regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score, recall_score, f1_score


log = LogisticRegression()     # Create a LogisticRegression object
log.fit(X_train, y_train)      # Train the model on training data
y_pred1 = log.predict(X_test)  # Predict labels for test data

print(accuracy_score(y_test,y_pred1))
# Using accuracy alone on an imbalanced dataset is dangerous because it can seem good but it is not.

print(precision_score(y_test,y_pred1))
print(recall_score(y_test,y_pred1))
print(f1_score(y_test,y_pred1))
# As we can see printed below, these other methods show very different scores
# So that is why it is important to evaluate the quality of a model using more than one score.
```

# Handling Imbalanced Data with SMOTE (Synthetic Minority Oversampling Technique)
For more information about this technique [click here](https://towardsdatascience.com/smote-fdce2f605729).

```python
# Using SMOTE to handle imbalanced data by generating synthetic samples for minority class.
# X_res and y_res are balanced (resampled) datasets after applying SMOTE.
# SMOTE helps improve model performance on imbalanced datasets.
from imblearn.over_sampling import SMOTE

X_res, y_res = SMOTE().fit_resample(X, y)   # Applying SMOTE to balance the dataset
y_res.value_counts()                        # Displaying the class distribution of the balanced dataset

```

```python
X_res
```

```python
# Splitting new balanced (resampled) Dataset
X_train,X_test,y_train,y_test=train_test_split(X_res,y_res,test_size=0.20,random_state=42)
```

```python
# Feature Scalling again, but now with the new balanced dataset 

sc = StandardScaler()                 # Create a StandardScaler object
X_train = sc.fit_transform(X_train)   # Fit on training data and transform features
X_test = sc.transform(X_test)         # Transform test features using the same scale

X_train
```

# Models Evaluation
1. Logistic Regression: 
A linear model used for binary or multiclass classification. It estimates the probability of a sample belonging to a certain class and uses a logistic function to make predictions.

2. SVC (Support Vector Classification): 
A classification algorithm that finds the hyperplane that best separates classes, aiming to maximize the margin between them. It can handle linear and non-linear separation with the use of kernel functions.

3. KNN (k-Nearest Neighbors): A non-parametric algorithm for classification and regression. It classifies data points by considering the class labels of their k closest neighbors in the feature space.

4. DT (Decision Tree): A tree-like model where each internal node represents a decision based on a feature, and each leaf node represents a class label. It's used for classification and regression tasks.

5. RF (Random Forest): An ensemble method that consists of multiple decision trees. Each tree is trained on a bootstrapped subset of the data, and the final prediction is the majority vote (classification) or average (regression) of individual tree predictions.

6. GBC (Gradient Boosting Classifier): An ensemble method that builds a strong classifier by iteratively adding weak learners (usually decision trees) while emphasizing the samples that were previously misclassified, resulting in improved performance.

In summary, these models vary in their underlying principles, how they make predictions, and their strengths and weaknesses. The choice of model depends on factors such as the nature of the problem, the characteristics of the data, and the desired performance metrics.

```python
 # Logistic Regression

log = LogisticRegression()     # Create a object
log.fit(X_train, y_train)      # Train the model on training data
y_pred1 = log.predict(X_test)  # Predict labels for test data

print(accuracy_score(y_test,y_pred1))
print(precision_score(y_test,y_pred1))
print(recall_score(y_test,y_pred1))
print(f1_score(y_test,y_pred1))
```

```python
# SVC(Support Vector Classification)
from sklearn import svm

svm = svm.SVC()
svm.fit(X_train,y_train)
y_pred2 = svm.predict(X_test)

print(accuracy_score(y_test,y_pred2))
print(precision_score(y_test,y_pred2))
print(recall_score(y_test,y_pred2))
print(f1_score(y_test,y_pred2))
```

```python
# KNN (k-Nearest Neighbors)
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
knn.fit(X_train,y_train)
y_pred3 = knn.predict(X_test)

print(accuracy_score(y_test,y_pred3))
print(precision_score(y_test,y_pred3))
print(recall_score(y_test,y_pred3))
print(f1_score(y_test,y_pred3))
```

```python
# DT (Decision Tree)
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
dt.fit(X_train,y_train)
y_pred4 = dt.predict(X_test)

print(accuracy_score(y_test,y_pred4))
print(precision_score(y_test,y_pred4))
print(recall_score(y_test,y_pred4))
print(f1_score(y_test,y_pred4))
```

```python
# RF (Random Forest)
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(X_train,y_train)
y_pred5= rf.predict(X_test)

print(accuracy_score(y_test,y_pred5))
print(precision_score(y_test,y_pred5))
print(recall_score(y_test,y_pred5))
print(f1_score(y_test,y_pred5))
```

```python
# GBC (Gradient Boosting Classifier)
from sklearn.ensemble import GradientBoostingClassifier

gbc = GradientBoostingClassifier()
gbc.fit(X_train,y_train)
y_pred6=gbc.predict(X_test)

print(accuracy_score(y_test,y_pred6))
print(precision_score(y_test,y_pred6))
print(recall_score(y_test,y_pred6))
print(f1_score(y_test,y_pred6))
```

# Random Forest model is the winner
It presented more accuracy and precision than the other models.

**Accuracy** = (True Positives + True Negatives) / (True Positives + True Negatives + False Positives + False Negatives)

**Precision** = True Positives / (True Positives + False Positives)

```python
# Now we are going to make it ready to be used in production.
# For that we must train our model with the entire balanced (resampled) Dataset

# Instead of using X_train from the splitted dataset, we are using the whole X_res

# Featuring scalling
X_res = sc.fit_transform(X_res)

# Train the RF model
rf.fit(X_res,y_res)
```

```python
# Import the joblib library for saving and loading models
import joblib

# Save the trained Random Forest model ('rf') to a file named 'churn_predict_rf_model'
joblib.dump(rf, 'churn_predict_rf_model')

# Load the saved model from the file 'churn_predict_rf_model' into a new variable 'rf_model'
rf_model = joblib.load('churn_predict_rf_model')
```

```python
# Calculate the correlation coefficients between numeric columns and the 'Exited' column
correlation = import_db.corr(numeric_only=True)['Exited']

# Sort the correlation values in descending order (greatest to smallest)
correlation = correlation.sort_values(ascending=False)

# Display the sorted correlation values
correlation
```

- Age has a moderate positive correlation (0.285) with 'Exited', indicating that older customers are slightly more likely to leave.
- 'Geography_Germany' has a positive correlation (0.173) with 'Exited', suggesting that customers from Germany have a slightly higher likelihood of leaving.
- 'Balance', 'EstimatedSalary', and other features have relatively weak correlations with - 'Exited', implying only a limited impact on customer attrition.
- 'IsActiveMember' and 'Gender_Male' have negative correlations, implying that active members and male customers are less likely to leave.

The correlations are generally not very strong, suggesting that customer attrition may be influenced by a combination of factors rather than a single dominant factor.
